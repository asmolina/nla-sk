### Problem 1.0 (Matrix norms) (8 pts)

1. For any norm $\|\cdot\|$ in finite dimentional space $\mathbb{C}^{N}$ we define dual norm $\|x\|^{'} = \sup\left\{\left|y^{\star} x\right|:\left\|y\right\|=1\right\}$, where ${\cdot}^{\star}$ denotes transpose complex-conjugate vector.

+ (2 pts). Prove that $\|\cdot\|^{'}$ is a norm.

**Proof.** 
In index notation:
$
\|x\|^{\prime}:=\sup _{\|y\|=1}\left|\left(y^*\right)^{\top} \cdot x\right| \equiv \sup _{\|y\|=1}\left| \sum_{i=1}^{N} \bar{y}_i x_i\right| \equiv \sup _{\|y\|=1}\left|\bar{y}_i x_i\right|
$
1) $\|x\|^{\prime}=\sup _{\|y\|=1}\left|\bar{y}_i x_i\right| \geqslant\left|\bar{y}_i x_i\right| \geqslant 0 \quad \forall x \in \mathbb{C}^N$ (because of a modulus)
$\|x\|^{\prime}=0 \Leftrightarrow \sup _{\|y\|=1}\left|\bar{y}_i x_i\right|=0 \Leftrightarrow \forall y:\|y\|=1 \hookrightarrow\left|\bar{y}_i x_i\right|=0$
$\Leftrightarrow x_i=0$ (since $\bar{y}_i \neq 0$ at least for one $i$ because of $\|y\|=1) \Leftrightarrow x=0$. In other words $\|x\|^{\prime} \geqslant  0$ and $\|x\|^{\prime}  =  0$ iff $x=0$.

2) Let's prove homogenity:
\begin{aligned}
&\|\alpha x\|^{\prime}=\sup _{\|y\|=1}\left|\overline{y_i}\left(\alpha x_i\right)\right| \\
&=\sup _{\|y\|=1} \sqrt{\left(\overline{y_i} \alpha x_i\right) \cdot \overline{\left(\bar{y}_i \alpha x_i\right)}} \\
&=\sup _{\|y\|=1} \sqrt{\alpha\left(\bar{y}_i x_i\right) \cdot \bar{\alpha}\left(\overline{y_i x_i}\right)} \\
&=\sup _{\|y\|=1}\left\{|\alpha| \cdot\left|\bar{y}_i x_i\right|\right\}= \\
&=|\alpha| \cdot \sup _{\|y\|=1}\left|\bar{y}_i x_i\right|=|\alpha|\|x\|^{\prime} \\
&
\end{aligned}

3) And triangle inequality:
\begin{aligned}
\left\|x^{1}+x^2\right\|^{\prime} &=\sup _{\|y\|=1}\left|\overline{y_i}\left(x_i^1+x_i^2\right)\right|=\\
&=\sup _{\|y\|=1}\left|\overline{y_i} x_i^1+\overline{y_i} x_i^2\right| \leqslant \\
& \leqslant \sup _{\|y\|=1} | \overline{y_i} x_i^{1} |+\sup _{\|y\|=1} | \overline{y_i} x_i^2 |=\\
&=\left\|x^1\right\|^{\prime}+\left\|x^2\right\|^{\prime}
\end{aligned}






+ (2 pts). Find dual norms for Chebyshev norm $\left\|x\right\|_{\infty} = \max\limits_{i=1,\dots, N}|x_{i}|$ and $l_1$ norm $\left\|x\right\|_1 = \sum\limits_{i=1}^{N}|x_{i}|$.

**Solution:**

a. 
\begin{aligned}

&\leqslant  \sup_{y_i : \max \left| y_i \right|=1} \sum_{i=1}^{N} \left| \bar{y}_i x_i\right|  \leqslant \\
&\leqslant  \sup_{y_i : \max \left| y_i \right|=1} \sum_{i=1}^{N}\left|\bar{y}_i\right|\left|x_i\right| \leqslant \\
&\leqslant \quad \max _{i=1, N}\left|y_i\right| \cdot  \sum_{i=1}^{N}\left|x_i\right|= \\
&=1 \cdot\sum_{i=1}^{N}\left|x_i\right|=\|x\|_{1} \\
\end{aligned}

We got $ \|x\|^{\prime}_{\infty} = \sup_{\|y\|_{\infty} \left| \right| = 1} \left|y^{\star} x\right| \leqslant \|x\|_{1} $. Let's provide $\widetilde{y}: \|\widetilde{y} \|_{\infty} =1$ at which the upper-bound for supremum is approached, i.e. $  \sup_{\|y\|_{\infty} = 1} \left|y^{\star} x\right| = \left|y^{\star} x\right| = \|x\|_{1} $. This will conclude he proof of $\|x\|^{\prime}_{\infty} = \|x\|_{1} $

Consider $\widetilde{y}_i = \text{sign } x_i$. Indeed, $ \|\widetilde{y}\|^{\prime}_{\infty} = \max \left| \widetilde{y}_i \right| = \max \left( 1, -1 \right) = 1$ and $\left|y^{\star} x\right| = \left| \sum_{i=1}^{N} \left( \text{sign } x_i \right) x_i \right| = \sum_{i=1}^{N}\left|x_i\right|=\|x\|_{1}$ 

b.

\begin{aligned}
\|x\|_1^{\prime} &=\sup _{y_i: \sum_{i=1}^N\left|y_i\right|=\sum_{i=1}^N\left|\bar{y_i}\right|=1}\left|\sum \bar{y}_i x_i\right| \leqslant \\

&\leqslant \quad \max _{i=\overline{1, N}}\left|x_i\right| \cdot \sup_{y_i : \sum_{i=1}^N\left|\bar{y_i}\right|=1} \sum_{i=1}^{N}\left|x_i\right|= \\
&=\max _{i=\overline{1, N}}\left|x_i\right|  \cdot 1=\|x\|_{\infty} \\
\end{aligned}

For any $x \in \mathbb{C}^{N}$ let's take $j (j=\overline{1, N}): \max _{i=\overline{1, N}}\left|x_i\right| = \left|x_j\right|$ and consider $\hat{y}=\left(0 0 \ldots 0 \operatorname{sign} x_j 0 \ldots 0\right)^{\top}$. $ \|\hat{y}\|_1=\left|\operatorname{sign} x_j\right|=1$ and
$$
\sum_{i=1}^N \overline{\hat{y}_i} \cdot x_i=\operatorname{sign} x_j \cdot x_j=\left|x_j\right|=\max _{i= \overline{1, N}}\left|x_i\right| \Rightarrow \forall x \in \mathbb{C}^N \quad\|x\|_1^{\prime}=\|x\|_{\infty}
$$



+ (2 pts). Find dual norm for $l_p$ norm $\left\|x\right\|_{p} = \left(\sum\limits_{i=1}^{N}|x_{i}|^{p}\right)^{1/p}$.

**Solution:**
\begin{aligned}
&\|x\|_p^{\prime}=\sup _{\|y\|_p=1}\left|\sum_{i=1}^N \bar{y}_i x_i\right| \leqslant \\
&\leqslant \sup _{\|y\|_p=1} \sum_{i=1}^N\left|\bar{y}_i x_i\right| \leqslant \\
&\leqslant \sup _{\left(\sum_{i=1}^N \left|\bar{y}_i \right|^p \right)^{\frac{1}{p}} = 1} \left(\sum_{i=1}^N\left|\bar{y}_i\right|^p\right)^{\frac{1}{p}}\left(\sum_{i=1}^N\left|x_i\right|^q\right)^{\frac{1}{q}} \leqslant \\
&\leqslant\left(\sum_{i=1}^N\left|x_i\right|^q\right)^{\frac{1}{q}}=\|x\|_q \\
&
\end{aligned}
where $q = \frac{p}{p-1}$ is given by HÃ¶lder's inequality.

Now among $y: \left| y_1 \right|^p + ... + \left| y_N \right|^p = 1 $ we need to find those which satisfies $\left| \bar{y}_1 x_1 ^p + ... +  \bar{y}_N x_N \right|^q = \left| x_1 \right|^q + ... + \left| x_N \right|^q $ or equivalently $y^* x = \| x \|_q$

Note that for $y_i=k x_i\left|x_i\right|^{q-2} \left( k \in \mathbb{R} \right) \hookrightarrow y^* x=\left|\sum_{i=1}^N \overline{y_i} x_i\right|=\left|\sum_{i=1}^N k \overline{x_i} x_i |x_i|^{q-2}\right|=k \sum_{i=1}^N |x_i|^q=k\|x\|_q^q$. We want to take $k$ so that $y^* x = k\|x\|_q^q = \| x \|_q \Rightarrow k=\|x\|_q^{1-q} $. 

Let's consider $\hat{y}_i=\|x\|_q^{1-q} \cdot\left|x_i\right|^{q-2} x_i=\|x\|_q^{-\frac{q}{p}} \cdot\left|x_i\right|^{q-2} x_i= \frac{\left| x_i \right| ^{q-2} x_i}{\|x\|_q^{q / p}} $

$\|\hat{y}\|_p^p=\sum\left|y_i\right|^p=\sum\left|\frac{x_i\left|x_i\right|^{q-2}}{\|x\|_q^{q / p}}\right|^p=\frac{\sum\left|x_i\right|^{(q-1) p}}{\|x\|_q^q}= \frac{\sum\left|x_i\right|^{\frac{q}{p} p}}{\|x\|_q^q}=\frac{\|x\|_q^q}{\|x\|_q^q}=1 $

$\Rightarrow \forall x \in \mathbb{C}^N \quad \exists \hat{y}_i=\frac{\left| x_i \right| ^{q-2} x_i}{\|x\|_q^{q / p}}:\|\hat{y}\|_p=1$ and $ y^* x =\|x\|_q^{1-q} $ $\Rightarrow \| x \|_p^{\prime} = \| x \|_q$



+ (2 pts). Find dual norm for weighted norm $\left\|x\right\|_{A} = \left(x^T A x\right)^{1/2}$ with positive-definite matrix $A$ and $x\in\mathbb{R}^{N}$.

**Solution**:
For positive definite and symmetric matrix $A$ Cholesky decomposition gives $L : A = L L^T$. $ A $ is invertible i.e. $\exists A^{-1}=\left(L L^{\top}\right)^{-1}=\left(L^{\top}\right)^{-1} L^{-1}: A^{-1} A=I \Rightarrow$

\begin{aligned}
&
\|x\|_{A}^{\prime}=\sup _{y^{\top} A y=1}\left|y^{\top} x\right|=\sup _{y^{\top} L L^{\top} y=1}\left|y^{\top} L L^{-1} x\right|=\left\{\begin{array}{l}
z^{\top}=y^{\top} L, \\
z=L^{\top} y
\end{array}\right\}=\sup _{(z,z)=1}\left|\left(z, L^{-1} x\right)\right| \leqslant \\
&\leqslant \sup _{(z, z)=1} \sqrt{|(z, z)|} \sqrt{\left|\left(L^{-1} x, L^{-1} x\right)\right|}=\sqrt{\left|\left(L^{-1} x, L^{-1} x\right)\right|}=\sqrt{ \left|  x^{\top} \left(L^{-1}\right)^{\top} L^{-1} x \right|}= \\
&=\sqrt{\left|x^{\top}\left(L L^{\top}\right)^{-1} x\right|}=\sqrt{x^{\top} A^{-1} x}
\end{aligned}

So we got the following inequality: $\|x\|_{A}^{\prime}=\sup _{\sqrt{y^{\top} A y}=1}\left|y^{\top} x\right| \leqslant \sqrt{x^{\top} A^{-1} x}$.

$\forall x \neq 0 $ $\exists \tilde{y}=\frac{A^{-1} x}{\sqrt{x^{\top}\left(A^{-1}\right)^{\top} x}}: \|\tilde{y}\|_A = 1 \text{ and } \hookrightarrow y^{\top} x =\sqrt{x^{\top}\left(A^{-1}\right)^{\top} x}$ 

Indeed, $\|\tilde{y}\|_A=\sqrt{\tilde{y}^{\top} A y}=\frac{1}{\sqrt{x^{\top}\left(A^{-1}\right)^T}x} \sqrt{x^{\top}\left(A^{-1}\right)^{\top} A A^{-1} x}=\frac{\sqrt{x^{\top}\left(A^{-1}\right)^{\top} x}}{\sqrt{x^{\top}\left(A^{-1}\right)^{\top} x}}=1$ and $y^{\top} x=\frac{\left(A^{-1} x\right)^{\top} x}{\sqrt{x^{\top}\left(A^{-1}\right)^{\top} x}}=\frac{x^{\top}\left(A^{-1}\right)^{\top} x}{\sqrt{x^{\top}\left(A^{-1}\right)^{\top} x}}=\sqrt{x^{\top}\left(A^{-1}\right)^{\top} x}$

$$

$$

1) calculate $m$ diagonal elements of $VV^T$: $\left[V V^{\top}\right]_{i i}=\sum_{\beta=1}^n v_{i \beta} v_{i \beta}=\sum_{\beta=1}^n v_{i \beta}^2  \rightarrow  O(m \cdot 2 r)$

1. We define approximate $\epsilon-$rank of matrix $A$ as $\text{rank}_{\epsilon}A = \min \left\{\text{rank}(X): \left\|A-X\right\|_{\max}\leq \epsilon\right\}$.

  
+ (6 pts). Let $f(x, y),\,x,y\in[0, 1]$ is analytic function with $\left.\partial_y^{(k)}f(x, y)\right|_{y=0}\leq M$ for all $k$. Show that approximate rank of matrix $A_{ij} = f(x_i, y_j)$ where $x_{i},\,i=1,\dots,p$, $y_{j},\,j=1,\dots,q$ are arbitrary points inside $[0, 1]^2$, is bounded independent of $p$, $q$ and the choice of points.


**Solution:** for analytic function $\varphi \left( y \right) = f \left(x, y \right)$ for given $x$ its Taylor series at point $y_i$ is convergent to $\varphi \left( y_i \right)$ in a neighbourhood of $y=0$.
$a_{i j}=f\left(x_i, y_i\right)=\left.\varphi(y)\right|_{x_i}=f\left(x_i, 0\right)+\sum_{k=1}^{\infty} \frac{\partial f_y^{(k)}\left(x_i, 0\right)}{k !} y_i^k $

Let us consider a matrix $X$:
$$x_{ij}=f\left(x_i, 0\right)+\sum_{k=1}^{N\left( \varepsilon \right)} \frac{\partial f_y^{(k)}\left(x_i, 0\right)}{k !} y_j^k $$


\begin{aligned}
\|A-X\|_{\max }=& \max _{i, j}\left|a_{i j}-x_{i j}\right|=\max _{i, j}\left|\sum_{k=N(\varepsilon)+1}^{\infty} \frac{\partial f_y^{(k)}\left(x_{i, 0}\right)}{k !} y_j^k\right| \leqslant \\
& \leq \max _{i, j} \sum_{k=N(\varepsilon)+1}^{\infty}\left|\frac{\partial f_y^{(k)}\left(x_i, 0\right)}{k !}\right| \cdot\left|y_j\right|^k=\\
&=\sum_{k=N(\varepsilon)+1}^{\infty} \frac{M}{k !} \cdot 1=M\left(\frac{1}{(N(\varepsilon)+1) !}+\frac{1}{(N(\varepsilon)+2) !}+\ldots\right)= \\
&=\frac{M}{(N(\varepsilon)+1) !}\left(1+\frac{1}{(N(\varepsilon)+2)}+\frac{1}{(N(\varepsilon)+2)(N(\varepsilon)+3)}+\ldots\right) \leqslant \\
&\leqslant \frac{M}{(N(\varepsilon)+1) !}\left(1+\frac{1}{N(\varepsilon)}+\frac{1}{N^2(\varepsilon)}+\ldots\right)= \\
&=\frac{M}{(N(\varepsilon)+1) !} \cdot \frac{1}{1-\frac{1}{N(\varepsilon)}}=\frac{M \cdot N(\varepsilon)}{(N(\varepsilon)+1) !(N(\varepsilon)-1)} \xrightarrow[N(\varepsilon) \to \infty]{} 0
\end{aligned}

It means that for given $\varepsilon$ we can find $N(\varepsilon)$ - number of terms in Taylors series which will allow to approximate matrix $A$ by matrix $X$ with $\varepsilon$ precision in terms of $\| \cdot \|_{\max}$ norm.

Matrix $X$ could be thought as a sum of $ N\left( \varepsilon \right) + 1$  rank-1 matrices. Indeed, 

$$x_{i j}=\sum_{k=0}^{N\left( \varepsilon \right)} \frac{\partial f_y^{(k)}\left(x_i, 0\right)}{k !} y_j^k=\sum_{k=0}^{N(\varepsilon)} \frac{1}{k !}\left(\begin{array}{c}
\partial f_y^{(k)}\left(x_1, 0\right) \\
\vdots \\
\partial f_y^{(k)}\left(x_p, 0\right)
\end{array}\right)\left(y_1^k \cdots y_q^k\right) \Rightarrow X =  \sum_{k=0}^{N\left( \varepsilon \right)} \frac{1}{k !} a_k b_k^{\top}, $$


where $ a_k=\left(\partial f_y^{(k)}\left(x_1, 0\right) \ldots \partial f_y^{(k)}\left(x_p, 0\right)\right)^{\top}, b_k^{\top}=\left(y_1^k \ldots y_q^k\right) $


Therefore the minimal rank of matrices approximating the given matrix $A_{p \times q}$ is bounded independently of its size $ \Rightarrow \operatorname{rank}_{\varepsilon} A \leqslant N\left( \varepsilon \right) + 1$

\end{array}\right]\left[\begin{array}{lll}
y_1^r \cdots y_1^z & \ldots & y_{q}^r \cdots y_q^z
\end{array}\right]
\end{equation}

totalling in $ 1 + N + N^2 + \cdots + N^{N\left( \varepsilon  \right)} $ terms. Thus $\operatorname{rank}_{\varepsilon} A  \leqslant 1 + N + N^2 + \cdots + N^{N\left( \varepsilon  \right)}$. + (2 pts). Prove that orthogonal transformations preserve $l_2$ norm, i.e., $\left\|x\right\|_2 = \left\|Ux\right\|_2$, where $U$ is an orthogonal matrix.


**Proof:** for any $x \in \mathbb{R}^{N}$: 
$$\left\|x\right\|_2 = \sqrt{ (x, x)} = \sqrt{ x^T x} = \sqrt{ x^T I x} = \sqrt{ x^T U^T U x} = \sqrt{ (Ux)^T U x}  = \sqrt{ (Ux, Ux)} =  \left\|Ux\right\|_2$$ 
Here were used $U^T = U^{-1}$ property for orthogonal matrices .


+ (4 pts). Use the result from the previous point to construct an algorithm that distinguishes different molecules. Implement it and run on the database. Produce a list of distinct molecules and visualise them.

**Solution:** for each molecule we will construct a representation in the form [($r_1$, $r_2$, $r_3$, $r_4$), ($\varphi_{12}$, $\varphi_{23}$, $\varphi_{34}$, $\varphi_{41}$)], where $r_i$ is a distance from the central atom A to the $i$-th atom, and $\varphi_{ij}$ is an angle between two adjacent edges ($\varphi_{12}$ + $\varphi_{23}$ + $\varphi_{34}$ + $\varphi_{41} = 2 \pi$). 

Angles and distances are preserved under arbitrary rotation in the context of this problem, so each distinct molecule will have its unique representation (probably, up to some cyclic permutation). That allows us to compare constructed representations for all of the molecules and find only unique ones.

```python
import numpy as np
np.seterr(divide='ignore', invalid='ignore')

def calculate_angle(vector):
    '''
    For given vector = (x, y) calculates the angle (in radians) between x-axis.
    Resulting angle is in range [0, 2*pi].
    '''
    x, y = vector[0], vector[1]
    alpha = np.arccos(x / (x**2 + y**2) ** 0.5)
    return alpha * (y>=0) + (2 * np.pi - alpha) * (y<0)


def calculate_relative_angles(dist_ang_array):
    '''
    Calculates relative angles for each molecule in the given array of representations.
    '''
    res = []
    for molecule in dist_ang_array:
        molecule = molecule[:, molecule[-1].argsort()[::-1]]
        angles = molecule[-1]
        angles = [x - y for x, y in zip(angles, np.append(angles[1:], angles[0] - 2*np.pi))]
        molecule[-1] = angles
        res.append(molecule)       
    return np.array(res)


def coincide_with_a_cyclic_permutation(molecule_1, molecule_2):
    """
    molecule_1 is compared to the cyclic permutations of molecule_2.
    In the worst scenario we will make N cyclic permutations, 
    where N is number of noncentral atoms 
    """
    for _ in range(molecule_2.shape[-1]):
        permutation_2 = np.append(molecule_2[:, 1:], np.reshape(molecule_2[:, 0], (2, 1)), axis=1)
        if np.allclose(molecule_1, permutation_2):
            return True
        else:
            molecule_2 = permutation_2     
    return False


def get_unique_molecules(molecules, dist_ang):
    unique_molecules = set(range(len(molecules)))

    for num_1 in range(len(molecules)):
        molecule_1 = dist_ang[num_1]
        same_as_molecule_1 = []
        
        if num_1 not in unique_molecules:
            continue

        for num_2, molecule_2 in enumerate(dist_ang):
            if coincide_with_a_cyclic_permutation(molecule_1, molecule_2):
                same_as_molecule_1.append(num_2)
            for item in same_as_molecule_1:
                unique_molecules.discard(item)
            unique_molecules.add(num_1)
    
    return unique_molecules
```

```python
# Calculate representations
distances = [(molecule**2).sum(axis=1)[1:] for molecule in molecules]
angles = [calculate_angle(molecule.T[:])[1:] for molecule in molecules]
dist_ang = np.array([[d, a] for d, a in zip(distances, angles)])
dist_ang = calculate_relative_angles(dist_ang)

# Find only unique among them
unique_molecules = get_unique_molecules(molecules, dist_ang)

# Vizualise the result
fig, ax = plt.subplots(1, len(unique_molecules), figsize=(2*len(unique_molecules), 2))
ax = ax.reshape(-1,)
for i, ax_i in zip(unique_molecules, ax):
    draw_molecule(ax_i, molecules[i])
plt.title('Distinct molecules')
plt.show()
```

+ (7 pts). Learn about [Kabsch algorithm](https://en.wikipedia.org/wiki/Kabsch_algorithm). Implement it and run on the database. Produce a list of distinct molecules and visualise them.

```python
def get_unique_molecules_kabsch(molecules):
    unique_molecules = set(range(len(molecules)))

    for num_1, molecule_1 in enumerate(molecules):
        if num_1 not in unique_molecules:
            continue
        
        same_as_molecule_1 = []
        for num_2, molecule_2 in enumerate(molecules):
            # Kabsh algorithm 
            u, s, vh = np.linalg.svd(molecule_1.T @ molecule_2, full_matrices=True)
            d = np.sign(np.linalg.det(u @ vh))
            R = vh.T @ np.diag([1, d]) @ u.T
            molecule_1_rotated = molecule_1 @ R.T
            
            if np.allclose(molecule_2, molecule_1_rotated):
                same_as_molecule_1.append(num_2)
            for item in same_as_molecule_1:
                unique_molecules.discard(item)
            unique_molecules.add(num_1)
    
    return unique_molecules

unique_molecules_kabsch = get_unique_molecules_kabsch(molecules)

# Vizualise the result
fig, ax = plt.subplots(1, len(unique_molecules_kabsch), figsize=(2*len(unique_molecules_kabsch), 2))
ax = ax.reshape(-1,)
for i, ax_i in zip(unique_molecules_kabsch, ax):
    draw_molecule(ax_i, molecules[i])
plt.title('Distinct molecules found by 2D Kabsch algorithm')
plt.show()
```

```python
def get_unique_molecules_3D_kabsch(molecules):
    unique_molecules = set(range(len(molecules)))

    for num_1, molecule_1 in enumerate(molecules):
        if num_1 not in unique_molecules:
            continue
            
        molecule_1 = np.hstack((molecule_1, np.zeros((molecule_1.shape[0],1))))    
        same_as_molecule_1 = []
        
        for num_2, molecule_2 in enumerate(molecules):
            molecule_2 = np.hstack((molecule_2, np.zeros((molecule_2.shape[0],1)))) 
            u, s, vh = np.linalg.svd(molecule_1.T @ molecule_2, full_matrices=True)
            d = np.sign(np.linalg.det(u @ vh))
            R = vh.T @ np.diag([1, 1, d]) @ u.T
            molecule_1_rotated = molecule_1 @ R.T
            
            if np.allclose(molecule_2, molecule_1_rotated):
                same_as_molecule_1.append(num_2)
            for item in same_as_molecule_1:
                unique_molecules.discard(item)
            unique_molecules.add(num_1)
    
    return unique_molecules

unique_molecules_3D_kabsch = get_unique_molecules_3D_kabsch(molecules)


# Vizualise the result
fig, ax = plt.subplots(1, len(unique_molecules_3D_kabsch), figsize=(2*len(unique_molecules_3D_kabsch), 2))
ax = ax.reshape(-1,)
for i, ax_i in zip(unique_molecules_kabsch, ax):
    draw_molecule(ax_i, molecules[i])
plt.title('Distinct molecules found by 3D Kabsch algorithm')
plt.show()
```

## Brief intro

TheÂ mainÂ componentÂ ofÂ [GAN](https://pytorch.org/tutorials/beginner/dcgan_faces_tutorial.html)Â isÂ GeneratorÂ network.Â GeneratorÂ isÂ suchÂ modelÂ thatÂ takesÂ asÂ inputÂ randomÂ vectorÂ andÂ producesÂ anÂ image.Â InÂ thisÂ task,Â weÂ willÂ workÂ withÂ theÂ alreadyÂ pre-trainedÂ generatorÂ $G$.Â HereÂ areÂ aÂ fewÂ thingsÂ youÂ needÂ toÂ knowÂ aboutÂ thisÂ model:



1.Â Â Â **ModelÂ input:**Â batchÂ ofÂ randomÂ vectorsÂ $z\simÂ N(0,1)$Â withÂ normallyÂ distributedÂ valuesÂ ofÂ sizeÂ **(Batch_Size,100)**.Â Here,Â Batch_SizeÂ isÂ a number ofÂ imagesÂ weÂ wantÂ toÂ generate

2.Â Â Â **ModelÂ output:**Â $\mathrm{images}Â =Â G(z)$Â -Â batchÂ ofÂ imagesÂ withÂ sizeÂ **(Batch_Size,3,64,64)**.Â HereÂ **3**Â -Â definesÂ numberÂ ofÂ imageÂ channelsÂ (3Â forÂ RGBÂ images),Â **64,Â 64**Â -Â widthÂ andÂ height.

3.Â Â Â **ModelÂ structure:**Â generatorÂ isÂ aÂ setÂ ofÂ sequentiallyÂ apppliedÂ functions.Â EachÂ functionÂ inputÂ isÂ theÂ outputÂ ofÂ the previousÂ one.Â InÂ otherÂ words,Â $G(z)Â =Â f_N(f_{N-1}...f_1(f_0(z)))$.Â $N$Â -Â definesÂ numberÂ ofÂ functionsÂ inÂ generator.



### Disentanglement

TheÂ goodÂ propertyÂ ofÂ GANÂ isÂ thatÂ generatorÂ inputÂ spaceÂ hasÂ aÂ setÂ ofÂ meaningfulÂ directionsÂ -Â movingÂ inputÂ vectorÂ alongÂ thisÂ directionsÂ produceÂ interpretableÂ changesÂ inÂ image.Â ForÂ example,Â weÂ canÂ changeÂ aÂ Pose,Â Age,Â ExpressionÂ orÂ evenÂ addÂ eyeglasses.Â YouÂ canÂ readÂ moreÂ aboutÂ itÂ onÂ [resource](https://genforce.github.io/interfacegan/).Â TheÂ mainÂ questionÂ isÂ howÂ toÂ findÂ suchÂ goodÂ directions.Â ThisÂ problemÂ isÂ calledÂ disentanglement.

WeÂ areÂ goingÂ toÂ findÂ meaningfulÂ directionsÂ usingÂ SVDÂ appliedÂ overÂ theÂ Jacobian matrixÂ ofÂ someÂ hiddenÂ outputÂ withÂ respectÂ toÂ modelÂ input. **i-th**Â hiddenÂ outputÂ ofÂ generatorÂ modelÂ is defined by formulaÂ $G_{i}(z)Â =Â f_i(f_{i-1}...f_1(f_0(z))))$,Â whereÂ $iÂ <Â N$.Â So,Â theÂ taskÂ couldÂ beÂ splittedÂ onÂ fourÂ mainÂ parts:

1.Â **(4Â pts)**Â Â Â WriteÂ theÂ functionÂ **Get_Hidden_Output**,Â whichÂ reproducesÂ formulaÂ $G_{i}(z)Â =Â f_i(f_{i-1}...f_1(f_0(z))))$

2.Â **(3Â pts)**Â Â Â CalculateÂ jacobian matrixÂ $J$Â ofÂ generatorÂ $G$Â forÂ severalÂ **i**Â values.Â ForÂ eachÂ ofÂ them:

```python
class Generator(nn.Module):
    def __init__(self, ngpu):
        super(Generator, self).__init__()
        self.ngpu = ngpu
        self.main = nn.Sequential(
            # input is Z, going into a convolution
            nn.ConvTranspose2d( nz, ngf * 8, 4, 1, 0, bias=False),
            nn.BatchNorm2d(ngf * 8),
            nn.ReLU(True),
            # state size. (ngf*8) x 4 x 4
            nn.ConvTranspose2d(ngf * 8, ngf * 4, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 4),
            nn.ReLU(True),
            # state size. (ngf*4) x 8 x 8
            nn.ConvTranspose2d( ngf * 4, ngf * 2, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf * 2),
            nn.ReLU(True),
            # state size. (ngf*2) x 16 x 16
            nn.ConvTranspose2d( ngf * 2, ngf, 4, 2, 1, bias=False),
            nn.BatchNorm2d(ngf),
            nn.ReLU(True),
            # state size. (ngf) x 32 x 32
            nn.ConvTranspose2d( ngf, nc, 4, 2, 1, bias=False),
            nn.Tanh()
            # state size. (nc) x 64 x 64
        )
    def get_function_output(self, inp, index=0):
      if index == 0:
        inp = inp[:,:,None,None]
      return self.main[index](inp)
    def get_number_of_layers(self):
      return len(self.main)

    def forward(self, input):
        return self.main(input[:,:,None,None])
```
